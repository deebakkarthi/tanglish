```python
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Read the actual and predicted POS tags from files
with open("tamil_testing_labels.txt", "r", encoding="utf-8") as file:
    actual_tags = file.read().splitlines()

with open("tamil_tags.txt", "r", encoding="utf-8") as file:
    predicted_tags = file.read().splitlines()
```

```python
y_actual=[]
for i in actual_tags:
    i=i.split()
    
    for j in i:
        y_actual.append(j)
```

```python
y_predict=[]
for i in predicted_tags:
    i=i.split()
    
    for j in i:
        y_predict.append(j)
```

```python
len(y_actual)
```

```python
len(y_predict)
```

```python

# Calculate accuracy
accuracy = accuracy_score(y_actual, y_predict)
print(f"Accuracy: {accuracy:.4f}")

# Calculate precision, recall, and F1 score
precision, recall, f1_score, _ = precision_recall_fscore_support(y_actual, y_predict, average='weighted')
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1_score:.4f}")

# Create a confusion matrix
conf_matrix = confusion_matrix(y_actual, y_predict)

# Plot the confusion matrix
plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=True, yticklabels=True)
plt.xlabel('Predicted Labels')
plt.ylabel('Actual Labels')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.metrics import precision_recall_fscore_support
import pandas as pd
```

```python
metrics_per_class = precision_recall_fscore_support(y_actual, y_predict, labels=list(set(y_actual)))
```

```python
per_class_df = pd.DataFrame({
    'POS Tag': list(set(y_actual)),
    'Precision': metrics_per_class[0],
    'Recall': metrics_per_class[1],
    'F1 Score': metrics_per_class[2],
    'Support': metrics_per_class[3]
})

# Display the per-class metrics
print(per_class_df)
```
